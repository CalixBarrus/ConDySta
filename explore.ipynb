```python
import pandas as pd
import os
```

```python
fd_report_directory = "data/experiments/fd_report_basic"

df_summary = pd.DataFrame()
for item in os.listdir(fd_report_directory):
    report = os.path.join(fd_report_directory, item)

    df = pd.read_csv(report)


    benchmark = "gpbench"
    col = "Leaks"

    if benchmark in item:
        
        df_summary[item] = df[col]
        if "App Name" not in df_summary.columns:
            df_summary["App Name"] = df["App Name"] 

ordered_columns = ["App Name", "ss-full-fd-default", "ss-per-scenario", "DEPTH_0_DA", "DISABLE_FIELD", "FULL_CONTEXT"]
ordered_columns = [col for ident in ordered_columns for col in df_summary.columns if ident in col]
df_summary[ordered_columns]

# df_summary.iloc[:, [5,4,2,0,3,1]]
```
